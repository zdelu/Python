Instrukcja range w python służy do generowania listy liczb.
Gdy chcemy, by dana czynność wykonała się z góry określiną liczbę razy stosujemy instrukcję range.

````python
for i in range(5):
    print("i = ", i)
````

````python
n=5
for i in range(n):
  print("*")
````

````python
n=5
for i in range(n):
  print("*", end="\n")
````

````python
n=5
for i in range(n):
  print("*", end="")
````

Domyślnie funkcja print() Pythona kończy się znakiem nowej linii. Programista może się zastanawiać, jak drukować bez znaku nowej linii, piszemy end="".
Funkcja print() Pythona zawiera parametr o nazwie „end”. Domyślnie wartość tego parametru to '\n', tj. znak nowej linii.

Znów mamy dwa trójkaty - z jednego musimy "skasować" jedną kolumnę.

````python
n=5
for i in range(n):
  for j in range(i+1):
    print(" ",end=" ")
  for j in range(i,n-1):
    print("*", end=" ")
  for j in range(i,n):
    print("*",end=" ")
  print()
````

````python
n=5
for i in range(n):
  for j in range(i,n):
    print(" ",end=" ")
  for j in range(i):
    print("*", end=" ")
  for j in range(i+1):
    print("*",end=" ")
  print()
for i in range(n):
  for j in range(i+1):
    print(" ",end=" ")
  for j in range(i,n-1):
    print("*", end=" ")
  for j in range(i,n):
    print("*",end=" ")
  print()
````

Jednak nasz diament nie ma spiczastych rogów, przyglądając się uważnie widzimy, że dwa wiersze  (środkowe mają taką samą szerokość) więc po prostu zmieniamy warunek pętli zewnętrznej. 

````python
n=10
for i in range(n-1):
  for j in range(i,n):
    print(" ",end=" ")
  for j in range(i):
    print("*", end=" ")
  for j in range(i+1):
    print("*",end=" ")
  print()
for i in range(n):
  for j in range(i+1):
    print(" ",end=" ")
  for j in range(i,n-1):
    print("*", end=" ")
  for j in range(i,n):
    print("*",end=" ")
  print()
````
